Calculating Euclidean Distance

```python
def euclidean_distance(X_data, single_point):
  return np.sqrt(np.sum((X_data - single_point)**2, axis=1))
```

Applying KNN-Classifier

```python
def knn_classifier(X_train, y_train, X_test_point, k):
    """
    k-nearest neighbors (KNN) classifier.

    Parameters:
    - X_train (ndarray): Training data features.
    - y_train (ndarray): Training data labels.
    - X_test_point (ndarray): A single test data point (features).
    - k (int): Number of neighbors to consider.

    Returns:
    - y_pred (int): Predicted label for the test data point.
    """

    distances = euclidean_distance(X_train, X_test_point)
    k_indices = np.argsort(distances)[:k]
    k_nearest_labels = y_train[k_indices]
    k_nearest_labels_int = k_nearest_labels.astype(int)
    prediction = np.argmax(np.bincount(k_nearest_labels_int))
    return prediction

prediction = knn_classifier(X_synthetic, y_synthetic, np.array(test_point), k)
print(prediction)
```

### Generating Sine and Cosine Data for Regression

Now, let's create synthetic data based on sine and cosine functions. We'll add some random noise to simulate real-world measurements, making it a suitable dataset for regression.

```python
X_sin_cos = np.linspace(0, 2 * np.pi, 100).reshape(-1, 1)

y_sin = np.sin(X_sin_cos) + np.random.normal(0, 0.1, X_sin_cos.shape)

y_cos = np.cos(X_sin_cos) + np.random.normal(0, 0.1, X_sin_cos.shape)

print("First 5 X values:\n", X_sin_cos[:50])
print("\nFirst 5 noisy sine values:\n", y_sin[:50])
print("\nFirst 5 noisy cosine values:\n", y_cos[:50])
```

### Visualize the Synthetic Sine and Cosine Data

Let's plot the generated sine and cosine data to see their distributions, including the added noise.

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(X_sin_cos, y_sin, label='Noisy Sine Data', alpha=0.7)
plt.title('Synthetic Sine Wave Data')
plt.xlabel('X')
plt.ylabel('sin(X) + noise')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.scatter(X_sin_cos, y_cos, label='Noisy Cosine Data', color='orange', alpha=0.7)
plt.title('Synthetic Cosine Wave Data')
plt.xlabel('X')
plt.ylabel('cos(X) + noise')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
def knn_regressor(X_train, y_train, X_test_point, k):
    """
    k-nearest neighbors (KNN) regressor.

    Parameters:
    - X_train (ndarray): Training data features.
    - y_train (ndarray): Training data labels.
    - X_test_point (ndarray): A single test data point (features).
    - k (int): Number of neighbors to consider.

    Returns:
    - y_pred (float): Predicted label (mean value) for the test data point.
    """
    distances = euclidean_distance(X_train, X_test_point)
    k_indices = np.argsort(distances)[:k]
    k_nearest_labels = y_train[k_indices]
    prediction = np.mean(k_nearest_labels)
    return prediction
```

```python
X_pred_sin = np.linspace(0, 2 * np.pi, 200).reshape(-1, 1)

y_pred_sin = np.array([knn_regressor(X_sin_cos, y_sin.flatten(), point, k) for point in X_pred_sin])

plt.figure(figsize=(10, 6))
plt.scatter(X_sin_cos, y_sin, label='Noisy Sine Data', alpha=0.7, color='blue')
plt.plot(X_pred_sin, y_pred_sin, label=f'KNN Regression (k={k})', color='red', linestyle='--')
plt.title('KNN Regression on Sine Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()

X_pred_cos = np.linspace(0, 2 * np.pi, 200).reshape(-1, 1)

y_pred_cos = np.array([knn_regressor(X_sin_cos, y_cos.flatten(), point, k) for point in X_pred_cos])

plt.figure(figsize=(10, 6))
plt.scatter(X_sin_cos, y_cos, label='Noisy Cosine Data', alpha=0.7, color='orange')
plt.plot(X_pred_cos, y_pred_cos, label=f'KNN Regression (k={k})', color='red', linestyle='--')
plt.title('KNN Regression on Cosine Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```
